### Создание таблицы матчапов

```python
df['player.hero_card_id'].unique()
```

```python
# Создание таблицы матчапов 9х9
matchup = pd.DataFrame(
    columns=[981,  754,  612,  494,   25, 1235,  326,  390,  798], 
    index=[ 981,  754,  612,  494,   25, 1235,  326,  390,  798]
)
```

```python
# Заполнение
for card_i in [ 981,  754,  612,  494,   25, 1235,  326,  390,  798]:
    for card_j in [ 981,  754,  612,  494,   25, 1235,  326,  390,  798]:
        matchup[card_j].loc[card_i] = df[df['player.hero_card_id']==card_j][df['opponent.hero_card_id']==card_i]['decision'].mean()
```

```python
# Столбцы - герой i
# Строки - герой j
# Значение - процент победы героя i против героя j
matchup
```

### Создание столбца броня+здоровье

```python
df['player.armor+hp'] = df['player.armor']+df['player.hp']
```

```python
df['player.armor+hp']
```

```python
armor_hp = []
decision = []
for val in df['player.armor+hp'].unique():
    decision.append(
        df[df['player.armor+hp'] == val]['decision'].mean()
    )
    armor_hp.append(val)
```

### График

```python
plt.scatter(armor_hp, decision)
plt.xlabel("Кол-во хп+брони")
plt.ylabel('Среднее значение decision для данного кол-ва хп+брони')
plt.title("Зависимость кол-ва хп+брони и процента победы")
plt.show()
```
